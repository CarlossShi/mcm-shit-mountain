```python
train_df = pd.read_csv(data_path + 'train_data.csv', dtype=str, nrows=100)
train_df
```

```python
train_userID = set(train_df.loc[:, 'userID'])
userID2idx = {_: str(i).zfill(7) for i, _ in enumerate(train_userID)}
```

```python
for i in range(len(train_df)):
    idx = userID2idx[train_df.userID[i]]
    aaa, bb, cc = idx[0:3], idx[3:5], idx[5:7]
    path = 'data_splitted_by_user_id/{}/{}/{}.csv'.format(aaa, bb, cc)
    if os.path.exists(path):
        train_df.loc[i:i].to_csv(path, mode='a', index=True, header=False)
    else:
        os.makedirs('data_splitted_by_user_id/{}/{}'.format(aaa, bb), exist_ok=True)
        train_df.loc[i:i].to_csv(path)
```

# Split All, Loop by `train_df`

```python
import tqdm
import pickle
```

```python
train_df = pd.read_csv(data_path + 'train_data.csv', dtype=str)
train_df
```

```python
if not os.path.exists(data_path + 'userID2idx.pickle'):
    train_df = pd.read_csv(data_path + 'train_data.csv', dtype=str)
    train_userID = set(train_df.loc[:, 'userID'])
    userID2idx = {_: str(i).zfill(7) for i, _ in enumerate(train_userID)}
    with open(data_path + 'userID2idx.pickle', 'wb') as handle:
        pickle.dump(userID2idx, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open(data_path + 'userID2idx.pickle', 'rb') as handle:
    userID2idx = pickle.load(handle)
```

```python
# 6h-16h execute time 
for i in tqdm.tqdm(range(len(train_df))):
    idx = userID2idx[train_df.userID[i]]
    aaa, bb, cc = idx[0:3], idx[3:5], idx[5:7]
    path = 'data_splitted_by_user_id/{}/{}/{}.csv'.format(aaa, bb, cc)
    if os.path.exists(path):
        train_df.loc[i:i].to_csv(path, mode='a', index=True, header=False)
    else:
        os.makedirs('data_splitted_by_user_id/{}/{}'.format(aaa, bb), exist_ok=True)
        train_df.loc[i:i].to_csv(path)
```
